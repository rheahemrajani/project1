3. Create a numpy array y containing the values for the natural logarithm function using the np.log(x) function. Create a numpy array  z  containing the values for the exponential function using the np.exp(x) function.

```python
log_arr = np.log(grid)
print(log_arr)

exp_arr = np.exp(grid)
print(exp_arr)
```

4. Use the plt.scatter(x,y) method for the  y  and  z  vectors to create two scatter plots of the points you've created.
5. Use plt.show() to create

```python
plt.scatter(grid, log_arr, color='blue', label='Log(x)', alpha=0.7)
plt.scatter(grid, exp_arr, color='red', label='Exp(x)', alpha=0.7)
plt.show()
```

6. Before the plt.show() call, add labels to the  x  and  y  axes using plt.xlabel(label) and plt.ylabel(label). Add a title to the graph using plt.title(title), like "Natural Log and Exponential Functions".





```python
plt.scatter(grid, log_arr, color='blue', label='Log(x)', alpha=0.7)
plt.scatter(grid, exp_arr, color='red', label='Exp(x)', alpha=0.7)
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Natural Log and Exponential Functions')
plt.show()
```

7.
That looks a lot better, but we need a legend. When you screate the scatter plots, add the argument label='Natural Log' or label='Exponential' to the .scatter method call. Before the plt.show() call, add a plt.legend(loc = 'lower right') method call, which creates a legend in the lower right.

```python
plt.scatter(grid, log_arr, color='blue', label='Log(x)', alpha=0.7)
plt.scatter(grid, exp_arr, color='red', label='Exp(x)', alpha=0.7)
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Natural Log and Exponential Functions')
plt.legend(loc='lower right')
plt.show()
```
